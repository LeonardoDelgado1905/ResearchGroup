### Creación de el trabajador de servientrega

```python
trabajador = Postman(np.random.randint(1, 100, 2))
```

### Creación de variables de temperatura para el templado simulado

```python
def funcTemp(t): #t representa el tiempo
    return 10*math.exp(-t/100)

time = 0
temperatura = funcTemp(time) #Temperatura inicial
```

- #### Función para nuevo orden dada una energía, Permutación inical de orden de visitas por parte del trabajador

```python
def changeOrder(prevOrder, energy):
    idxToChange = np.random.choice(numUsuarios, energy, replace = False)
    trans = np.random.permutation(idxToChange)
    newOrder = prevOrder    
    for i in range(energy):
        newOrder[idxToChange[i]] = prevOrder[trans[i]] #permuta entre si los índices obtenidos
    return newOrder

ordenEntregas = np.random.permutation(numUsuarios)
```

### Calcular Energía

```python
def energyCalc(order):
    cost = 0
    currPoint = trabajador
    for i in order:
        first = currPoint.coor
        cost += np.linalg.norm(usuarios[i].coor - currPoint.coor)
        currPoint = usuarios[i]
    return cost

```
